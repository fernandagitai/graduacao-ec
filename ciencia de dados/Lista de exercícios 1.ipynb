### 1. Descreva como um cientista de dados aproveita das técnicas da Descoberta de Aprendizado em Base de dados (KDD). Ainda, explique como um especialista atua junto nesse processo da ciência de dados e como um conhecimento aprofundado do cientista pode auxiliar nesse processo. (1,5 pontos)

O cientista de dados trabalha em função de extrair informações de um conjunto de dados, seja para o âmbito dos negócios ou de pesquisa. A Descoberta de Aprendizado em Base de dados (KDD) descreve um conjunto de etapas que servem de modelo para o cientista de dados exercer o seu trabalho a partir de um problema.

A expertise do cientista de dados neste processo entra em como cada etapa será realizada, baseando-se no seu problema específico e no seu conjunto de dados. Saber quais e como os dados serão selecionados (etapa de seleção), definir qual o formato dos dados é melhor (etapa de pré-processamento e transformação) e saber lidar com os demais problemas das etapas restantes depende do conhecimento aprofundado do cientista de dados e não apenas da descrição das técnicas propostas pelo KDD.

### 2. Explique o que é Big Data e como a ciência de dados se relaciona com ela. Existe uma diferença significativa do que é considerado Big Data hoje com uma análise de dados grandes de 2005, por exemplo? Se sim, descreva como e cite dois exemplos práticos. (1,5 pontos)

Apesar de serem dois termos usados corriqueiramente como semelhantes, o termo *Big Data* se refere ao problema de coleta, processamento e gerenciamento de dados, muito importante para trabalhar com um volume de dados de alta complexidade e variedade, enquanto o termo *Ciência de dados* lida com o problema de solucionar a modelagem dos dados, ou seja, de obter um conhecimento dos dados. Ambos os termos estão diretamente ligados, visto que ciência de dados utiliza ferramentas e técnicas específicas para extrair valor do *Big Data*, o que envolve a coleta, processamento, análise e interpretação de grandes conjuntos de dados.

Comparando, desta forma, a análise de *Big Data* agora com a feita em 2005, notamos uma grande diferença visto o aumento da importância dos dados e consequentemente do volume de dados gerados diariamente. As  fontes de dados agora são muito mais diversas e complexas, incluindo dados gerados por redes sociais, sensores, dispositivos móveis, transações financeiras, registros de saúde eletrônicos e dentre vários outras fontes.

Um exemplo prático para isso é a análise de vendas, na qual era processado um volume de dados na faixa de centenas de milhares de registros quaisquer, hoje temos empresas como a Amazon que lida com bilhões de transações a níveis mundiais, as quais servem como base para identificações de padrões no mercado e alavancamento das vendas.

Outro exemplo a ser comparado com o de 2005 é o monitoramento de saúde pública no país, onde anteriormente os dados eram coletados manualmente pelos funcionários de hospitais e clínicas e hoje se diferenciam por envolver a análise de dados de múltiplas fontes, incluindo redes sociais, registros eletrônicos de saúde, responsáveis estas por identificar rapidamente surtos de doenças e tomar medidas para contê-las.

<h1> Carregando a base de dados</h1>

Para esse exercício, a base de dados utilizada será a Iris. A qual já vem na biblioteca sklearn. As células a seguir carregam a base de dados e mostram alguns gráficos. Neste momento, o objetivo é apenas dar uma ideia do comportamento da base de dados para o aluno e não será cobrado nenhuma questão para desenhar gráficos ou carregar a base de dados.

```python
from  sklearn import  datasets
from sklearn.model_selection import train_test_split

#Carrega a base de dados iris direto do pacote sklearn
iris=datasets.load_iris()

#Define nossos x e y que serão utilizados nos exercicios seguintes
x = iris.data
y = iris.target
```

```python
import matplotlib.pyplot as plt
from sklearn import datasets

X = iris.data[:, :2] 
y = iris.target

x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

plt.figure(2, figsize=(8, 6))
plt.clf()

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA


fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
```

Utilizando a biblioteca scikit-learn complete o código inserindo os classificadores solicitados.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

random_state = 42


decision_tree = DecisionTreeClassifier(random_state=random_state)
random_forest = RandomForestClassifier(random_state=random_state)
svm =  SVC(random_state=random_state)
```

<h1>Complete o código a seguir  (1,0 ponto)</h1>

Depois de criar os classificadores, o próximo passo é avaliar o desempenho deles. Para isso, vamos utilizar a técnica de validação cruzada com K = 10 para cada um deles. Assim poderemos comparar o desempenho de cada um.
    
Os seguintes links podem ser úteis:

https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html

```python
from sklearn.model_selection import cross_val_score

k = 10

scores_decision_tree = cross_val_score(decision_tree, X, y, cv=k)
scores_random_forest = cross_val_score(random_forest, X, y, cv=k)
scores_svm = cross_val_score(svm, X, y, cv=k)

print("Score decision tree: ", scores_decision_tree.mean())
print("Score random forest: ", scores_random_forest.mean())
print("Score svm: ", scores_svm.mean())
```

<h1> Hiperparâmetro </h1>

Os hiperparâmetros são parâmetros ajustáveis que permitem controlar o processo de treinamento do modelo, alguns algoritmos têm seu desempenho severamente afetado dependendo do hiperparâmetro escolhido. Para escolhermos o melhor hiperparâmetro para o nosso algoritmo, podemos utilizar as  métricas de avaliação e medir seu desempenho para cada hiperparâmetro e, então, escolher o que tiver a melhor performance. 

<h1>Complete o código a seguir  (1,5 pontos)</h1>

Complete o código inserindo o KNN com um K variante e avalie o desempenho para cada valor de K.
    
Os seguintes links podem ser úteis:

https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

for i in range(2, 30, 2):
    knn = KNeighborsClassifier(n_neighbors=i) # inserir o KNN com o valor do hiperparâmetro K igual ao i do laço for
    scores = cross_val_score(knn, X, y, cv=i) # inserir a validação cruzada para cada modelo gerado 
    print("K =", i, "\t\t Score =", scores.mean())
```

<h1> Clusterização </h1>

O clustering ou análise de agrupamento de dados é o conjunto de técnicas de prospecção de dados (data mining) que visa fazer agrupamentos automáticos de dados segundo o seu grau de semelhança. Diferente da classificação, a clusterização não tem a label que nós utilizamos como y até agora. Então, para avaliar ela, precisamos de uma nova métrica que utilize apenas os dados do nosso X. Uma das métricas mais utilizadas é o Silhouette.

<h1> Complete o código a seguir (1,5 pontos)</h1>

Complete o código a seguir inserindo o KMeans dentro do laço for  e o silhouette_score para avaliar o desempenho do KMeans para cada número de cluster.

Os seguintes links podem ser  úteis:

https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html
https://scikit-learn.org/stable/modules/clustering.html#silhouette-coefficient

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics


for i in range(2, 11):
    kmeans_model = KMeans(n_clusters=i).fit(X) #inserir o  KMeans
    labels = kmeans_model.labels_
    score = metrics.silhouette_score(X, labels) #colocar o calculo do  silhouette_score
    print("K =", i, "\t\t Score =", score)
```
