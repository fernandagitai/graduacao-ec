```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

# 1a questão

```python
img = cv2.imread('pato.jpg', 0)
```

```python
# Transformada Rápida de Fourier (FFT)
f = np.fft.fft2(img)

# Calculando o espectro de magnitude
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))

plt.figure(figsize=(10,10))
plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Imagem de Entrada')
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Espectro de Magnitude')
plt.show()
```

```python
real = np.real(f)
imag = np.imag(f)
mod = np.abs(f)

plt.figure(figsize=(20,10))
plt.subplot(131),plt.imshow(real, cmap = 'gray')
plt.title('Parte Real')
plt.subplot(132),plt.imshow(imag, cmap = 'gray')
plt.title('Parte Imaginária')
plt.subplot(133),plt.imshow(mod, cmap = 'gray')
plt.title('Módulo')
plt.show()
```

```python
# Calculando a Transformada de Fourier
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)

# Calculando o módulo da Transformada de Fourier sem o shift
f_abs = np.abs(f)

# Separando a parte real e imaginária sem o shift
f_real = np.real(f)
f_imag = np.imag(f)

# Separando a parte real e imaginária com o shift
f_real_shift = np.real(fshift)
f_imag_shift = np.imag(fshift)

# Calculando o módulo da Transformada de Fourier com o shift
f_abs_shift = np.abs(fshift)
```

```python
plt.figure(figsize=(20,10))

plt.subplot(231),plt.imshow(np.log(1 + f_abs), cmap = 'gray')
plt.title('Módulo da Transformada sem Shift')

plt.subplot(234),plt.imshow(np.log(1 + f_abs_shift), cmap = 'gray')
plt.title('Módulo da Transformada com Shift')

plt.subplot(232),plt.imshow(f_real, cmap = 'gray')
plt.title('Parte Real sem Shift')

plt.subplot(235),plt.imshow(f_real_shift, cmap = 'gray')
plt.title('Parte Real com Shift')

plt.subplot(236),plt.imshow(f_imag, cmap = 'gray')
plt.title('Parte Imaginária sem Shift')

plt.subplot(233),plt.imshow(f_imag_shift, cmap = 'gray')
plt.title('Parte Imaginária com Shift')

plt.show()
```

# 4a questão

```python
img = cv2.imread('pato.jpg', 0)
```

```python
# Calculando a Transformada de Fourier
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)

# Criando o filtro passa-baixa
rows, cols = img.shape
crow, ccol = rows//2, cols//2
mask = np.zeros((rows, cols), np.uint8)
r = 40
cv2.circle(mask, (ccol, crow), r, 255, -1)

# Aplicando o filtro na frequência
fshift = fshift * mask

# Transformada inversa
f_ishift = np.fft.ifftshift(fshift)
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)

plt.figure(figsize=(10,10))
plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Imagem Original')
plt.subplot(122),plt.imshow(img_back, cmap = 'gray')
plt.title('Imagem com Filtro Passa-Baixa')
plt.show()
```

```python
img = cv2.imread('pato.jpg', 0)
```

```python
# Calculando a Transformada de Fourier
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)

# Criando o filtro passa-alta
rows, cols = img.shape
crow, ccol = rows//2, cols//2
mask = np.ones((rows, cols), np.uint8) * 255
r = 40
cv2.circle(mask, (ccol, crow), r, 0, -1)

# Aplicando o filtro na frequência
fshift = fshift * mask

# Realizando a transformada inversa
f_ishift = np.fft.ifftshift(fshift)
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)

plt.figure(figsize=(10,10))
plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Imagem Original')
plt.subplot(122),plt.imshow(img_back, cmap = 'gray')
plt.title('Imagem com Filtro Passa-Alta')
plt.show()
```
