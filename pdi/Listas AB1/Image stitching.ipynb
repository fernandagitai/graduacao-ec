```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Lendo os dados
ic1 = cv2.cvtColor(cv2.imread("ic1.jpg"), cv2.COLOR_BGR2RGB)
ic2 = cv2.cvtColor(cv2.imread("ic2.jpg"), cv2.COLOR_BGR2RGB)
ic3 = cv2.cvtColor(cv2.imread("ic3.jpg"), cv2.COLOR_BGR2RGB)
ic4 = cv2.cvtColor(cv2.imread("ic4.jpg"), cv2.COLOR_BGR2RGB)

fig, ax = plt.subplots(2,2, figsize=(10,7))

ax[0,0].imshow(ic1)
ax[0,0].set_title('Parte 1')

ax[0,1].imshow(ic2)
ax[0,1].set_title('Parte 2')

ax[1,0].imshow(ic3)
ax[1,0].set_title('Parte 3')

ax[1,1].imshow(ic4)
ax[1,1].set_title('Parte 4')

plt.show()
```

```python
imgs = [ic1,ic2,ic3,ic4]

stitchy=cv2.Stitcher.create()
(dummy,output)=stitchy.stitch(imgs)
output = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(15,10))
plt.title("Imagem panorama", fontsize=16)
plt.imshow(output)
plt.show()
```
