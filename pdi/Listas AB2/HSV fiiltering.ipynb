# Plotar os canais HSV separadamente

```python
def plot_HSV(img):
  img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)


  fig, axs = plt.subplots(1,3, figsize=(15,6))

  hue = img[:,:,0]
  saturation = img[:,:,1]
  value = img[:,:,2]

  axs[0].imshow(hue, cmap="gray")
  axs[0].set_title("Filtering Hue")
  axs[1].imshow(saturation, cmap="gray")
  axs[1].set_title("Filtering saturation")
  axs[2].imshow(value, cmap="gray")
  axs[2].set_title("Filtering value")
  
  plt.show()
```

```python
plot_HSV(img1)
```

```python
plot_HSV(img2)
```

```python
plot_HSV(img3)
```

```python
plot_HSV(img4)
```

# Normalizar cada imagem convertendo para ponto flutuante e dividindo por 255.0

```python
def normalizar_img(img):
  img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
  img = img.astype(float)

  fig, axs = plt.subplots(1,3, figsize=(15,6))

  H = img[:,:,0]
  min_value = np.min(H);  max_value = np.max(H)
  H_norm = ((H - min_value) / (max_value - min_value)) * 255
  H_norm = H_norm.astype(np.uint8)

  S = img[:,:,1]
  min_value = np.min(S);  max_value = np.max(S)
  S_norm = ((S - min_value) / (max_value - min_value)) * 255
  S_norm = S_norm.astype(np.uint8)

  V = img[:,:,2]
  min_value = np.min(V);  max_value = np.max(V)
  V_norm = ((V - min_value) / (max_value - min_value)) * 255
  V_norm = V_norm.astype(np.uint8)

  axs[0].imshow(H_norm, cmap="gray")
  axs[0].set_title("Filtering Hue")
  axs[1].imshow(S_norm, cmap="gray")
  axs[1].set_title("Filtering saturation")
  axs[2].imshow(V_norm, cmap="gray")
  axs[2].set_title("Filtering value")
  
  plt.show()
```

```python
normalizar_img(img1)
```

```python
normalizar_img(img2)
```

```python
normalizar_img(img3)
```

```python
normalizar_img(img4)
```
