# Plotar os canais RGB separadamente

```python
def plot_rgb(img):
  fig, axs = plt.subplots(1,3, figsize=(15,6))

  red = img[:,:,0]
  green = img[:,:,1]
  blue = img[:,:,2]

  axs[0].imshow(red, cmap="gray")
  axs[0].set_title("Filtrando a cor vermelha")
  axs[1].imshow(green, cmap="gray")
  axs[1].set_title("Filtrando a cor verde")
  axs[2].imshow(blue, cmap="gray")
  axs[2].set_title("Filtrando a cor azul")
  
  plt.show()
```

```python
plot_rgb(img1)
```

```python
plot_rgb(img2)
```

```python
plot_rgb(img3)
```

```python
plot_rgb(img4)
```

# Normalizar cada imagem convertendo para ponto flutuante e dividindo por 255.0

```python
def normalizar_img(img):
  img = img.astype(float)

  fig, axs = plt.subplots(1,3, figsize=(15,6))

  red = img[:,:,0]
  min_value = np.min(red);  max_value = np.max(red)
  red_norm = ((red - min_value) / (max_value - min_value)) * 255
  red_norm = red_norm.astype(np.uint8)

  green = img[:,:,1]
  min_value = np.min(green);  max_value = np.max(green)
  green_norm = ((green - min_value) / (max_value - min_value)) * 255
  green_norm = green_norm.astype(np.uint8)

  blue = img[:,:,2]
  min_value = np.min(blue);  max_value = np.max(blue)
  blue_norm = ((blue - min_value) / (max_value - min_value)) * 255
  blue_norm = blue_norm.astype(np.uint8)

  axs[0].imshow(red_norm, cmap="gray")
  axs[0].set_title("Cor vermelha normalizada")
  axs[1].imshow(green_norm, cmap="gray")
  axs[1].set_title("Cor verde normalizada")
  axs[2].imshow(blue_norm, cmap="gray")
  axs[2].set_title("Cor azul normalizada")
  
  plt.show()
```

```python
normalizar_img(img1)
```

```python
normalizar_img(img2)
```

```python
normalizar_img(img3)
```

```python
normalizar_img(img4)
```
