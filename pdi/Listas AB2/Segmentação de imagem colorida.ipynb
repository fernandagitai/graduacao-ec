```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/chips.png")
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

plt.imshow(img)
plt.show()
```

```python
r, g, b = cv2.split(img)
```

```python
# Defina o limite superior para o canal vermelho e os limites inferiores para os canais verde e azul
red_threshold = 200
green_threshold = 100
blue_threshold = 100

# Crie máscaras para cada canal de cor
red_mask = r > red_threshold
green_mask = g < green_threshold
blue_mask = b < blue_threshold

# Combine as máscaras para obter a máscara final
final_mask = np.logical_and(red_mask, np.logical_and(green_mask, blue_mask))

# Aplique a máscara à imagem original
filtered_red = np.zeros_like(img)
filtered_red[final_mask] = img[final_mask]

# Salve a imagem filtrada
plt.imshow(filtered_red)
plt.title("Filtro da cor vermelha")
plt.show()
```

```python
red_threshold = 100
green_threshold = 100
blue_threshold = 200

red_mask = r < red_threshold
green_mask = g < green_threshold
blue_mask = b > blue_threshold

final_mask = np.logical_and(blue_mask, np.logical_and(green_mask, red_mask))

filtered_blue = np.zeros_like(img)
filtered_blue[final_mask] = img[final_mask]

plt.imshow(filtered_blue)
plt.title("Filtro da cor azul")
plt.show()
```

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

lower_green = np.array([40, 80, 80])
upper_green = np.array([60, 255, 255])

green_mask = cv2.inRange(hsv_img, lower_green, upper_green)

green_image = cv2.bitwise_and(img, img, mask=green_mask)

plt.imshow(green_image)
plt.title("Filtro da cor verde")
plt.show()
```

```python
lower_yellow = np.array([90, 90, 90])
upper_yellow = np.array([95, 255, 255])

yellow_mask = cv2.inRange(hsv_img, lower_yellow, upper_yellow)

yellow_image = cv2.bitwise_and(img, img, mask=yellow_mask)

plt.imshow(yellow_image)
plt.title("Filtro da cor amarela")
plt.show()
```

```python
lower_orange = np.array([110, 90, 90])
upper_orange = np.array([120, 255, 255])

orange_mask = cv2.inRange(hsv_img, lower_orange, upper_orange)

orange_image = cv2.bitwise_and(img, img, mask=orange_mask)

plt.imshow(orange_image)
plt.title("Filtro da cor laranja")
plt.show()
```
